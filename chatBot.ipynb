```python
import pandas as pd
import json
import re
import numpy as np 
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
```

```python
with open('data.json') as file:
  data=json.load(file)
```

```python
data
```

```python
trainning_example=[]
trainning_label=[]

for example in data['data']:
  for pattern in example['patterns']:
    trainning_example.append(pattern)
    trainning_label.append(example['label'])
```

#preprocessing

```python
def cleanText(pattern):
  result=[]
  for sent in pattern:
    sent=re.sub('\W'," ",sent)
    sent=sent.lower()
    result.append(sent)

  return result
```

```python
trainning=cleanText(trainning_example)
trainning
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
trainnig_labels=le.fit_transform(trainning_label)
```

```python
labels=to_categorical(trainnig_labels)
```

```python
labels
```

```python
# Tokenizer which give every word unique value 
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

## convert tWord to vector

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts(trainning)
sequances=tokenizer.texts_to_sequences(trainning)
# every word get a value which represent the importance of this word 
```

```python
print(trainning)
```

```python
print(sequances)
```

```python
# want to make sure that the size of the input all the same so we will make padding
```

#padding


```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
paded_sequences=pad_sequences(sequances,maxlen=15)
```

```python
print(paded_sequences)
```

# Models

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Embedding,GlobalAveragePooling1D
```

```python
model=Sequential()
model.add(Embedding(500,10,input_length=15))
model.add(GlobalAveragePooling1D())
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(8,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(paded_sequences,labels,epochs=450)
```

#test

```python
inp='hello'
sent=pad_sequences(tokenizer.texts_to_sequences([inp]),maxlen=15)
sent
```

```python
result=np.argmax(model.predict(sent))
```

```python
final_result=le.inverse_transform(np.array(result).reshape(1))
```

## interface

```python
def userChat():
  print('# note # to close the app write Quit ###')
  print('#_______________________________________')
  while True:

    print('user : ',end='')
    inp=input()

    inp=re.sub('\W'," ",inp).lower()

    if inp.lower() == 'quit':
      break

    sent=pad_sequences(tokenizer.texts_to_sequences([inp]),maxlen=15)
    result=np.argmax(model.predict(sent))
    final_result=le.inverse_transform(np.array(result).reshape(1))


    for label in data['data']:
      if label['label']==final_result:
        print('Chat Bot : ',np.random.choice(label['responses']))
  
```
